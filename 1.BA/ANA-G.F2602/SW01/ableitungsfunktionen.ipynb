# Ableitungsfunktionen

Funktionen die zu jeder Stelle $x$ die Steigung der ursprünglichen Funktion angeben
- Ausgangsfunktion: $f(x)$
- Ableitungsfunktion $f'(x)$

**Bedeutung**
- $f'(x) > 0$: $f$ steigt
- $f'(x) < 0$: $f$ fällt
- $f'(x) = 0$: Tangente ist horizontal zur x-Achse

## Ableitungsregeln

### Konstantenregel

- **Regel**: $f(x)=c$ → $f'(x)=0$

- **Beispiel**: $f(x)=5$ → $f'(x)=0$

### Potenzregel

- **Regel**: $f(x)=x^n$ → $f'(x)=nx^{n-1}$

- **Beispiel**: $f(x)=x^5$ → $f'(x)=5x^4$

### Faktorregel

- **Regel**: $f(x)=c \times g(x)$ → $f'(x)=c \times g'(x)$

- **Beispiel**: $f(x)=4x^3$ → $f'(x)=4 \times 3x^2 = 12x^2$

### Summenregel

- **Regel**: $f(x)=g(x)+h(x)$ → $f'(x)=g'(x)+h'(x)$

- **Beispiel**: $f(x)=x^3+2$ → $f'(x)=3x^2+0=3x^2$

### Produktregel

- **Regel**: $f(x)=g(x) \times h(x)$ → $f'(x)=g'(x)h(x)+g(x)h'(x)$

- **Beispiel**: $f(x)=x \times x^2$ → $f'(x)= 1 \times x^2 + x \times 2x = x^2+2x^2 = 3x^2$

### Quotientregel

- **Regel**: $f(x)= \frac{g(x)}{h(x)}$ → $f'(x)= \frac{g'(x)h(x)-g(x)h'(x)}{h(x)^2}$




## Regel fur $x^n$

Fur Potenzen gilt:

$$
\frac{d}{dx}(x^n)=n\cdot x^{n-1}
$$

Damit folgt:

- $(x^3)'=3x^2$
- $(x^5)'=5x^4$
- $\left(\frac{1}{x}\right)'=(x^{-1})'=-x^{-2}=-\frac{1}{x^2}$


```python
import numpy as np
import matplotlib.pyplot as plt

# Funktionen
def p1(x):
    return x**3


def p5(x):
    return x**5


def r(x):
    return 1 / x


# Ableitungen
def p1_abl(x):
    return 3 * x**2


def p5_abl(x):
    return 5 * x**4


def r_abl(x):
    return -1 / x**2
```

```python
# x-Bereiche
x_poly = np.linspace(-2, 2, 400)

# Bei 1/x darf x nicht 0 sein, deshalb zwei getrennte Bereiche
x_r_left = np.linspace(-2, -0.1, 300)
x_r_right = np.linspace(0.1, 2, 300)

fig, axs = plt.subplots(1, 3, figsize=(15, 4))

# Plot 1: x^3 und Ableitung
axs[0].plot(x_poly, p1(x_poly), label="p1(x) = x^3")
axs[0].plot(x_poly, p1_abl(x_poly), label="p1'(x) = 3x^2")
axs[0].axhline(0, color="black", linewidth=0.8)
axs[0].axvline(0, color="black", linewidth=0.8)
axs[0].grid(True, alpha=0.3)
axs[0].set_xlabel("x")
axs[0].set_ylabel("y")
axs[0].set_title("x^3 und Ableitung")
axs[0].legend()

# Plot 2: x^5 und Ableitung
axs[1].plot(x_poly, p5(x_poly), label="p5(x) = x^5")
axs[1].plot(x_poly, p5_abl(x_poly), label="p5'(x) = 5x^4")
axs[1].axhline(0, color="black", linewidth=0.8)
axs[1].axvline(0, color="black", linewidth=0.8)
axs[1].grid(True, alpha=0.3)
axs[1].set_xlabel("x")
axs[1].set_ylabel("y")
axs[1].set_title("x^5 und Ableitung")
axs[1].legend()

# Plot 3: 1/x und Ableitung
axs[2].plot(x_r_left, r(x_r_left), label="r(x) = 1/x")
axs[2].plot(x_r_right, r(x_r_right))
axs[2].plot(x_r_left, r_abl(x_r_left), label="r'(x) = -1/x^2")
axs[2].plot(x_r_right, r_abl(x_r_right))
axs[2].axhline(0, color="black", linewidth=0.8)
axs[2].axvline(0, color="black", linewidth=0.8)
axs[2].grid(True, alpha=0.3)
axs[2].set_xlabel("x")
axs[2].set_ylabel("y")
axs[2].set_title("1/x und Ableitung")
axs[2].legend()

plt.tight_layout()
plt.show()
```

## Kurzinterpretation

- Bei $x^3$ und $x^5$ ist die Ableitung wieder ein Polynom, mit kleinerem Exponenten.
- Die Funktion $1/x$ und ihre Ableitung sind bei $x=0$ nicht definiert.
- Fur $x^n$ gilt allgemein die Potenzregel $n\cdot x^{n-1}$.
