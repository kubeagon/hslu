# Tangentensteigung

Die Tangentensteigung ist die Steigung der Tangente an eine Kurve an einem bestimmten Punkt. Sie beschreibt also die momentane Änderungsrate der Funktion an einem bestimmten Punkt

- **Gegeben**: Eine Funktion $f(x)$ und eine feste Stelle $x_0$

- **Gesucht**: Steigung der Tangente zur Funktion $f$ an der Stelle $x_0$ und $P(x_0+h/f(x_0+h))$

- **Formel** h nähert sich 0 (keine Differenz zwischen den Punkten)

$$
f'(x_0)= \lim_{h \to 0} \frac{f(x_0+h)-f(x_0)}{h}
$$

Um einen Graph erstellen zu können wird ein sehr kleines $h$ gewählt. Beispielsweise ist $1 \times 10^{-5}$ oder $1 \times 10^{-6}$
 passend

### Beispiel

Tangente zur Funktion $f(x)=x^2$ an der Stelle $x_0=2$

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

x0 = 2
h = 1e-5

def tangentensteigung(f, x0, h):
    return (f(x0 + h) - f(x0 - h)) / (2 * h)

m = tangentensteigung(f, x0, h)
print(f"Die Steigung der Tangente an der Stelle x0 = {x0} ist: {m}")

t = f(x0) - m * x0  # y-Achsenabschnitt
x = np.linspace(x0 - 1, x0 + 1, 200)
y = m * x + t

plt.figure(figsize=(7, 4))
plt.plot(x, y, label="Tangente an f(x) bei x0")
plt.scatter([x0], [f(x0)], color="red", zorder=3, label="Punkt auf der Kurve")
plt.axhline(0, color="black", linewidth=0.8)
plt.axvline(0, color="black", linewidth=0.8)
plt.grid(True, alpha=0.3)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Graph der Tangente an f(x) bei x0")
plt.legend()
plt.show()
```
