# Geradensteigung

**Gegeben**
- Punkt $P_1 = (x_1, y_1)$
- Punkt $P_2 = (x_2, y_2)$

**Gesucht**
- Die Steigung $m$ der Geraden durch beide Punkte

$$
m = \frac{\Delta y}{\Delta x} = \frac{y_2-y_1}{x_2-x_1}\ = \frac{y_1-y_2}{x_1-x_2}
$$

```python
def slope(p1, p2):
    x1, y1 = p1
    x2, y2 = p2

    if x2 == x1:
        return None  # vertical line: no finite slope

    return (y2 - y1) / (x2 - x1)
```

```python
# Example from the slide: P1(-1, 2), P2(2, 4)
p1 = (-1, 2)
p2 = (2, 4)

m = slope(p1, p2)

if m is None:
    print("Vertical line: slope is undefined.")
else:
    print(f"Steigung m = {m} (expected: 2/3)")
```

```python
import numpy as np
import matplotlib.pyplot as plt

x1, y1 = p1
x2, y2 = p2

if m is None:
    # Vertical line x = x1
    ys = np.linspace(min(y1, y2) - 2, max(y1, y2) + 2, 200)
    xs = np.full_like(ys, x1)
else:
    xs = np.linspace(min(x1, x2) - 2, max(x1, x2) + 2, 200)
    b = y1 - m * x1
    ys = m * xs + b

plt.figure(figsize=(7, 4))
plt.plot(xs, ys, label="Gerade durch P1 und P2")
plt.scatter([x1, x2], [y1, y2], color="red", zorder=3, label="Punkte")
plt.axhline(0, color="black", linewidth=0.8)
plt.axvline(0, color="black", linewidth=0.8)
plt.grid(True, alpha=0.3)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Graph der Geraden")
plt.legend()
plt.show()
```
