# Geradengleichung

Eine Geradengleichung ist die Gleichung welches die Gerade mathematisch beschreibt.

## Gerade mit der Steigung m und dem $y$-Achsenabschnitt $t$

$$
y(x) = m * x + t
$$

- $m$ ist die Steigung
    - ($m$ > 0) → Gerade geht nach oben rechts
    - ($m$ < 0) → Gerade geht nach unten rechts 
    - ($m$ = 0) → Gerade ist horizontal zur x-Achse

- $t$ ist wo y-Achsenabschnitt (Der Punkt, an dem eine Gerade die y-Achse schneidet)
    - Der $y$-Wert bei ($x$ = 0)

- $x$ ist der Eingabewert
    - $y$($x$) ist der zugehörige Funktionswert


```python
def y(x, m, t):
    return m * x + t
```

```python
# Example Steigung 3, y-Achsenabschnitt 2, x = 4
print(y(4, 3, 2))
```

```python
import numpy as np
import matplotlib.pyplot as plt

def line(x, m, t):
    return m * x + t

m, t = 3, 2
# Create many evenly spaced x-values from -5 to 5 (200 points),
# so matplotlib can draw a full line instead of just a single point.
x = np.linspace(-5, 5, 200)
y = line(x, m, t)

x0 = 4
y0 = line(x0, m, t)

plt.figure(figsize=(7, 4))
plt.plot(x, y, label="Gerade: y = 3x + 2")
plt.scatter([x0], [y0], color="red", label=f"Punkt ({x0}, {y0})")
plt.axhline(0, color="black", linewidth=0.8)
plt.axvline(0, color="black", linewidth=0.8)
plt.grid(True, alpha=0.3)
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

## Gerade durch einen festen Punkt mit Steigung $m$

**Ausgangslage:**

Gegeben sind die Steigung $m$ und ein fester Punkt $P_0(x_0, y_0)$ auf der Geraden.  
Gesucht ist die Geradengleichung.

Für jeden Punkt $P(x,y)$ auf derselben Geraden gilt über die Steigung:

$$
m = \frac{\Delta y}{\Delta x} = \frac{y-y_0}{x-x_0}
$$

Nach $y$ umformen:

$$
y-y_0 = m(x-x_0)
$$

$$
y(x) = m(x-x_0) + y_0
$$

Das ist die **Punkt-Steigungs-Form**. Ausmultipliziert erhält man:

$$
y(x) = mx + (y_0 - mx_0)
$$

Damit ist die Form $y=mx+t$ erreicht, mit: (Gesucht ist eine Konstante)

$$
t = y_0 - mx_0
$$



Also:  
- $m$ bestimmt die Steigung.  
- $(x_0,y_0)$ ist ein bekannter Punkt der Geraden.  
- $t$ ist der y-Achsenabschnitt.


```python
def line(x, m, x0, y0):
    return m * x + (y0 - m * x0)
```

```python
# Example Steigung 3, Punkt (4, 14)
print(line(4, 3, 4, 14))
```

```python
import numpy as np
import matplotlib.pyplot as plt

def line(x, m, x0, y0):
    return m * x + (y0 - m * x0)

m = 3
x0 = 3
y0 = 8

# Create many evenly spaced x-values from -5 to 5 (200 points),
# so matplotlib can draw a full line instead of just a single point.
x = np.linspace(-5, 5, 200)
y = line(x, m, x0, y0)

t = y0 - m * x0  # y-intercept

plt.figure(figsize=(7, 4))
plt.plot(x, y, label=f"Gerade: y = {m}x + {t}")
plt.scatter([x0], [y0], color="red", label=f"Punkt ({x0}, {y0})")
plt.axhline(0, color="black", linewidth=0.8)
plt.axvline(0, color="black", linewidth=0.8)
plt.grid(True, alpha=0.3)
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

## Gerade durch die feste Punkte $P_1(x_1/y_1)$ und $P_2(x_2/y_2)$

Für alle Punkte $P(x/y)$ auf der Geraden

$$
m = \frac{\Delta y}{\Delta x} = \frac{y_2-y_1}{x_1-x_2} = \frac{y-y_1}{x-x_1}
$$

Solve $y$ using the slope definition

$$
y(x) = \frac{y_2-y_1}{x_2-x_1} * (x - x_1) + y_1
$$

```python
def line(x, x1, y1, x2, y2):
    m = (y2 - y1) / (x2 - x1)
    return m * (x - x1) + y1
```

```python
print(line(4, 1, 2, 3, 5))  # Example: line through points (1, 2) and (3, 5) evaluated at x=4
```

```python
import numpy as np
import matplotlib.pyplot as plt

def line(x, x1, y1, x2, y2):
    m = (y2 - y1) / (x2 - x1)
    return m * (x - x1) + y1

x1 = 3
x2 = 5
y1 = 8
y2 = 2

# Create many evenly spaced x-values from -5 to 5 (200 points),
# so matplotlib can draw a full line instead of just a single point.
x = np.linspace(-5, 5, 200)
y = line(x, x1, y1, x2, y2)

plt.figure(figsize=(7, 4))
plt.plot(x, y, label=f"Gerade durch Punkte ({x1}, {y1}) und ({x2}, {y2})")
plt.scatter([x1, x2], [y1, y2], color="red", label=f"Punkte ({x1}, {y1}) und ({x2}, {y2})")
plt.axhline(0, color="black", linewidth=0.8)
plt.axvline(0, color="black", linewidth=0.8)
plt.grid(True, alpha=0.3)
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
