# Sekantensteigung

Ein Sekantensteigung ist die Steigung zwischen zwei Punkten auf einer Kurve

## Mit zwei feste Stellen

- **Gegeben**: Eine Funktion $f(x)$ und zwei feste Stellen $x_1$ und $x_2$

- **Gesucht**: Steigung der Sekante durch die Punkte $P_1(x_1/f(x_1))$ und $P_2(x_2/f(x_2))$

- **Formel** ($f(x)$ wird nur als $f(x_1)$ und $f(x_2)$ aufgeführt in der Formel)

Mathematisch gleich da invers (Zähler und Nenner werden mit -1 multipliziert)

$$
m = \frac{\Delta y}{\Delta x} = \frac{f(x_2)-f(x_1)}{x_2-x_1} = \frac{f(x_1)-f(x_2)}{x_1-x_2}
$$

### Beispiel

Annahme $f(x) = x^2$ mit Stelle $x_1=1$ und Stelle $x_2=3$

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = 1
x2 = 3

def sekantensteigung1(f, x1, x2):
    return (f(x2) - f(x1)) / (x2 - x1)

def f(x):
    return x**2

y1 = f(x1)
y2 = f(x2)

m = sekantensteigung1(f, x1, x2)
print(f"Die Steigung der Sekante durch die Punkte ({x1}, {y1}) und ({x2}, {y2}) ist: {m}")

# Anwendung y-Achsenabschnitt y(x) = mx + t → t = y(x) - mx
# Hier verwenden wir den Punkt P1 (x1, y1) um t zu berechnen
t = y1 - m * x1
x = np.linspace(x1 - 1, x2 + 1, 200) # x-Werte von etwas vor x1 bis etwas nach x2, 200 Punkte für eine glatte Linie
y = m * x + t

plt.figure(figsize=(7, 4))
plt.plot(x, y, label="Gerade durch P1 und P2")
plt.scatter([x1, x2], [y1, y2], color="red", zorder=3, label="Punkte")
plt.axhline(0, color="black", linewidth=0.8)
plt.axvline(0, color="black", linewidth=0.8)
plt.grid(True, alpha=0.3)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Graph der Sekante durch P1 und P2")
plt.legend()
plt.show()
```

## Eine feste Stelle und eine zweite mit definierten Abstand

- **Gegeben**: Eine Funktion $f(x)$, eine feste Stelle $x_0$ und eine zweite Stelle mit Abstand $h$

- **Gesucht**: Steigung der Sekante durch die beiden Punkte $P_0(x_0/f(x_0))$ und $P(x_0+h/f(x_0+h))$

- **Formel**

$$
m = \frac{\Delta y}{\Delta x}=\frac{f(x_0+h)-f(x_0)}{h}
$$

### Beispiel

Annahme $f(x)=x^2$ mit $x_0=2$ und $h=5$

```python
import numpy as np
import matplotlib.pyplot as plt

x0 = 2
h = 5

def f(x):
    return x**2

def sekantensteigung2(f, x0, h):
    return (f(x0 + h) - f(x0)) / h

m = sekantensteigung2(f, x0, h)
print(f"Die Steigung der Sekante durch die Punkte ({x0}, {f(x0)}) und ({x0 + h}, {f(x0 + h)}) ist: {m}")

x1 = x0
x2 = x0 + h
y1 = f(x0)
y2 = f(x0 + h)

t = y1 - m * x0  # y-Achsenabschnitt
x = np.linspace(x0 - 1, x0 + h + 1, 200)
y = m * x + t

plt.figure(figsize=(7, 4))
plt.plot(x, y, label="Gerade durch P1 und P2")
plt.scatter([x1, x2], [y1, y2], color="red", zorder=3, label="Punkte")
plt.axhline(0, color="black", linewidth=0.8)
plt.axvline(0, color="black", linewidth=0.8)
plt.grid(True, alpha=0.3)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Graph der Sekante durch P1 und P2")
plt.legend()
plt.show()
```
